```python
import pandas as pd
```

```python
# Load the Excel file containing the customer call list into a DataFrame
df=pd.read_excel(r"C:\Users\PC\Documents\doc\Customer Call List.xlsx")
```

```python
df
```

```python
# Remove duplicate rows from the DataFrame 'df' to ensure data integrity
df=df.drop_duplicates()
```

```python
# Remove the column 'Not_Useful_Column' from the DataFrame 'df' as it is not needed for analysis
df=df.drop(columns='Not_Useful_Column')
```

```python
# Remove specified leading and trailing characters ('...', '/', and '_') from the 'Last_Name' column in the DataFrame 'df'
df['Last_Name']=df['Last_Name'].str.strip(".../_")
```

```python
# Remove all non-alphanumeric characters from the 'Phone_Number' column in the DataFrame 'df'
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
# Convert all entries in the 'Phone_Number' column to string format in the DataFrame 'df'

df['Phone_Number']=df['Phone_Number'].apply(lambda x:str(x))
```

```python
# Format the 'Phone_Number' column by converting it to the 'XXX-XXX-XXXX' format in the DataFrame 'df'

df['Phone_Number']=df['Phone_Number'].apply(lambda x:x[0:3]+'-'+x[3:6]+'-'+x[6:10])
```

```python
# Remove occurrences of 'Na--' from the 'Phone_Number' column in the DataFrame 'df'

df['Phone_Number']=df['Phone_Number'].str.replace('Na--','')
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
# Split the 'Address' column into 'Street_address', 'State', and 'Zip_code' using a comma as the delimiter

df[['Street_address','State','Zip_code']]=df['Address'].str.split(',', n=2, expand=True)
```

```python
# Replace 'Yes' with 'Y' and 'No' with 'N' in the 'Do_Not_Contact' column
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')

# Replace 'Yes' with 'Y' and 'No' with 'N' in the 'Paying Customer' column
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
```

```python
# Replace all missing (NaN) values in the DataFrame 'df' with empty strings
df=df.fillna('')
```

```python
# Iterate through the DataFrame 'df' and remove rows where the 'Do_Not_Contact' column equals 'Y'

for x in df.index :
    if df.loc[x,'Do_Not_Contact']=='Y':
        df.drop(x,inplace=True)
```

```python
# Iterate through the DataFrame 'df' and remove rows where the 'Phone_Number' column is empty

for x in df.index :
    if df.loc[x,'Phone_Number']=='':
        df.drop(x,inplace=True)
```

```python
# Reset the index of the DataFrame 'df' and drop the old index
df=df.reset_index(drop=True)
```
